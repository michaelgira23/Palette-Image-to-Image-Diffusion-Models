```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image

lr_image = '../matter-density-superresolution/test/lr/1-110.png'
hr_image = '../matter-density-superresolution/test/hr/1-110.png'

def pil_loader(path):
    return Image.open(path).convert('RGB')

# Load images using numpy
lr = pil_loader(lr_image)
hr = pil_loader(hr_image)

# Plot images
plt.figure(figsize=(10, 10))
plt.subplot(1, 2, 1)
plt.imshow(lr, cmap='gray')
plt.title('Low Resolution')
plt.subplot(1, 2, 2)
plt.imshow(hr, cmap='gray')
plt.title('High Resolution')


plt.show()
```
